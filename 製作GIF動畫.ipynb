```
from PIL import Image, ImageDraw

def create_bouncing_ball_gif(filename='bouncing_ball.gif'):
    """
    生成一個簡單的紅色彈跳球 GIF 動畫。
    """
    # 設定畫布大小與顏色
    width, height = 400, 300
    bg_color = (255, 255, 255) # 白色背景
    ball_color = (255, 0, 0)   # 紅色球
    ball_radius = 30

    frames = []

    # 動畫參數
    total_frames = 40
    # 模擬球的垂直運動 (簡單的正弦波或拋物線邏輯)
    # 這裡使用簡單的線性上下移動作為範例

    # 球的 X 軸位置 (固定在中間)
    x = width // 2

    # 生成每一幀
    for i in range(total_frames):
        # 創建一個新的空白畫布
        img = Image.new('RGB', (width, height), bg_color)
        draw = ImageDraw.Draw(img)

        # 計算球的 Y 軸位置 (讓它上下移動)
        # 使用簡單的數學來模擬彈跳：從上到下再到上
        if i < total_frames // 2:
            # 下落 (0 到 height-radius)
            progress = i / (total_frames // 2)
            y = int(ball_radius + (height - 2 * ball_radius) * progress)
        else:
            # 上升
            progress = (i - total_frames // 2) / (total_frames // 2)
            y = int((height - ball_radius) - (height - 2 * ball_radius) * progress)

        # 畫圓 (左上角 x, y, 右下角 x, y)
        draw.ellipse(
            (x - ball_radius, y - ball_radius, x + ball_radius, y + ball_radius),
            fill=ball_color,
            outline=(0, 0, 0)
        )

        # 將這一幀加入列表
        frames.append(img)

    # 儲存為 GIF
    # duration: 每一幀停留的時間 (毫秒)
    # loop: 0 表示無限循環
    frames[0].save(
        filename,
        save_all=True,
        append_images=frames[1:],
        optimize=False,
        duration=50,
        loop=0
    )
    print(f"成功生成 GIF 檔案: {filename}")

if __name__ == "__main__":
    # 檢查是否有安裝 Pillow
    try:
        create_bouncing_ball_gif()
    except ImportError:
        print("錯誤: 請先安裝 Pillow 函式庫。")
        print("請在終端機執行: pip install pillow")
```
